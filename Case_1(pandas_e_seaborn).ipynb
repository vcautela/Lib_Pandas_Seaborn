```python
import pandas as pd
import seaborn as sns
```

```python
aluguel_casa = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/CaseBibliotecas(pandas e seaborn)/houses_rent.xlsx")
```

```python
aluguel_casa.head()
```

```python
#função head() e tail() mostra por padrão as 5 primeiras/últimas linhas
#ao inserir o parametro '3' na função tail() ou head(), você muda a qnt de linhas mostradas
aluguel_casa.tail(3)
```

```python
aluguel_casa = aluguel_casa.rename(columns={"parking spaces":"parking_spaces","rent amount":"rent_amount","property tax":"property_tax","fire insurance":"fire_insurance"})
```

```python
aluguel_casa.head()
```

```python
aluguel_casa.dtypes
```

```python
# substituindo tipos de dado em colunas selecionadas com '.astypes' e verificando o tipo dado com 'dtypes'
aluguel_casa = aluguel_casa.astype({"rent_amount":"float", "property_tax":"float", "fire_insurance":"float", "total":"float"})
aluguel_casa.dtypes
```

```python
# função info(): dtypes + robusta.
aluguel_casa.info()
```

```python
# começamos com a análise descritiva
# selecionando apenas colunas numéricas
aluguel_casa[["area", "rooms", "bathroom", "parking_spaces", "rent_amount", "property_tax", "fire_insurance", "total"]].describe()
```

Pela análise acima, é possível verificar insights importantes o dataframe, como por exemplo:

a média (mean) é o valor da soma dos valores das linhas pela quantidade de linhas, o desvio padrão(std) representa o quão distante da média.

ao verificar por exemplo, que na coluna "area" temos um desvio padrão muito maior que a média, é possível concluir que temos valores muito diversos em nosso dataframe, uma grande diferenciação entre os dados. (apartamentos muito grandes e apartamentos muito pequenos)

25% dos meus dados estão até 58m² (1º quartil)

a mediana da área é de 100m²

```python
#filtrando a partir das seguintes condições:Aluguel até $4000, aceita animais, 2 ou 3 quartos
aluguel_casa = aluguel_casa[(aluguel_casa["total"]<=4000) & (aluguel_casa["animal"]=="accept") & ((aluguel_casa["rooms"]==2) | (aluguel_casa["rooms"]==3))]
aluguel_casa.head()
```

```python
aluguel_casa.describe()
```

```python
# outra forma de filtrar a partir do '.query()'
aluguel_casa = aluguel_casa.query("(total <= 4000 and animal =='accept') and (rooms == 2 or rooms == 3)")
aluguel_casa.head()
```

```python
# agregando a coluna ('.groupby') cidade para relacionar a média de valores totais.
# reset_index() - deixa num formato de dataframe mais apresentável
agg_city_price = aluguel_casa.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
agg_bath_room_price = aluguel_casa.groupby(["bathroom", "rooms"])["total"].agg("mean").reset_index()
agg_bath_room_price
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="rooms", y="total", color="green");
ax.set(xlabel = "qtd quartos", ylabel = "média do preço");
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y="total", color="green");
ax.set(xlabel = "qtd banheiros", ylabel = "média do preço");
```
